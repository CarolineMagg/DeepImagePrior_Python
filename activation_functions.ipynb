# Create visualization of activation functions

* Leaky Relu
* Relu
* Swish
* Mish

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
%config Completer.use_jedi = False
```

```python
x = np.arange(-1.02, 1.02, 0.01)
```

```python
relu_y = tf.keras.activations.relu(x).numpy()
leaky_relu_y = tf.keras.layers.LeakyReLU(alpha=0.2)(x).numpy()
mish_y = x * tf.math.tanh(tf.math.softplus(x)).numpy()
swish_y = tf.keras.activations.swish(x).numpy()
```

```python
plt.plot(x, [0]*len(x), "k-")
plt.plot([0]*len(x), x, "k-")
plt.plot(x, relu_y, "b--", label="relu")
plt.plot(x, leaky_relu_y, "r--", label="leaky relu")
plt.plot(x, swish_y, "g-", label="swish")
plt.plot(x, mish_y, "m-", label="mish")
plt.legend()
plt.xlim(left=-1.02, right=1.02)
plt.ylim(bottom=-1.02, top=1.02)
plt.grid()
plt.show()
```
